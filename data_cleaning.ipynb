```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mvps = pd.read_csv('mvps.csv')
```

```python
mvps
```

```python
mvps = mvps[['Player', 'Year', "Pts Won", "Pts Max", "Share"]]
```

```python
mvps.head()
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players['Unnamed: 0']
del players['Rk']
```

```python
players
```

```python
players['Player'].head(50)
```

```python
players['Player'] = players['Player'].str.replace("*", "", regex=False) 
```

```python
def single_row(df): 
    if df.shape[0]==1: 
        return df
    else:
        row = df [df ["Team"] == "TOT"]
        row["Team"] = df.iloc[-1,:]["Team"]
        return row

players =players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined
```

```python
combined[combined["Pts Won"] > 0]
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

```python
teams = pd.read_csv("team_stats.csv")
```

```python
teams.head(10)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(10)
```

```python
teams.columns = teams.columns.str.strip()
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False) 
```

```python
teams
```

```python
teams["Team"].unique()
```

```python
combined["Team"].unique()
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Team"].map(nicknames)
```

```python
combined.head(5)
```

```python
del combined["Awards"]
```

```python
combined["Team"] = combined["Team"].str.strip()
teams["Team"] = teams["Team"].str.strip()

combined["Team"] = combined["Team"].str.lower()
teams["Team"] = teams["Team"].str.lower()

# Se ainda houver problemas de acentuação ou codificação:
combined["Team"] = combined["Team"].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
teams["Team"] = teams["Team"].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
stats = combined.merge(teams, how="inner", on=["Team", "Year"])
```

```python
stats
```

```python
del stats['Unnamed: 0']
```

```python
stats["GB"] = stats["GB"].str.replace("—", "0")
```

```python
stats = stats.apply(pd.to_numeric, errors="ignore")
```

```python
stats.dtypes
```

```python
stats.to_csv("player_mvp_stats.csv")
```
